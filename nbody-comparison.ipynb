# Comparison with $N$-body simulations

```python
rng_seed = 5281
```

```python
def nbody_vs_secular(N, m, a, t=50_000, save=False):
    simpair = SimulationPair(N, simulation_time=t, rng_seed=rng_seed)
    simpair.add(m=m, a=a)
    simpair.get_ll_systems()
    simpair.get_ll_solutions()
    simpair.plot_nbody_and_secular(save=save)
    return
```

The following are some comparisons between predictions from second-order secular theory and direct $N$-body integration of the system from various points in the parameter space in Fig. 5.
